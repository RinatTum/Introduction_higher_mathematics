```python
import matplotlib.pyplot as plt
```

```python
def draw_circle(r=1, x0=0, y0=0, color='C1'):
    angle = np.linspace(0, 2*np.pi, 360)
    
    x = np.add(r*np.cos(angle), x0)
    y = np.add(r*np.sin(angle), y0)
    
    plt.plot(x, y, color=color, linewidth=1)
    plt.gca().set_aspect('equal')
    
draw_circle(x0=1, y0=2)
```

```python
def draw_ellipse(a=2, b=1, x0=0, y0=0, color='C1'):
    angle = np.linspace(0, 2*np.pi, 360)
    
    x = np.add(a*np.cos(angle), x0)
    y = np.add(b*np.sin(angle), y0)
    
    plt.plot(x, y, color=color, linewidth=1)
    plt.gca().set_aspect('equal')
    
draw_ellipse(x0=1, y0=2)
```

```python
def draw_hyperbola(a=1, b=1, x_range=[-5, 5], color='C1'):
    epsilon = np.finfo(float).eps
    f0 = epsilon+a
    _x1 = np.linspace(x_range[0], -f0, (x_range[1]-x_range[0])*100)
    _x2 = np.linspace(f0, x_range[1], (x_range[1]-x_range[0])*100)
    
    f = lambda x: np.sqrt((x**2/a**2) - 1)/b
    
    _y1 = np.array([f(v) for v in _x1])
    _y2 = np.array([f(v) for v in _x2])
    _y3 = np.array([-f(v) for v in _x1])
    _y4 = np.array([-f(v) for v in _x2])
    
    plt.plot(_x1, _y1, color=color, linewidth=1)
    plt.plot(_x2, _y2, color=color, linewidth=1)
    plt.plot(_x1, _y3, color=color, linewidth=1)
    plt.plot(_x2, _y4, color=color, linewidth=1)
    plt.gca().set_aspect('equal')

draw_hyperbola()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
def draw_parallel_planes(A=0, B=0, C=1, D=0, z_offset=0):
    fig = figure()
    ax = Axes3D(fig)
    X = np.arange(-5, 5, 1)
    Y = np.arange(-5, 5, 1)
    X, Y = np.meshgrid(X, Y)
    Z = (A*X + B*Y + D)/C
    Z2 = (A*X + B*Y + D)/C + z_offset
    ax.plot_wireframe(X, Y, Z, color='green')
    ax.plot_wireframe(X, Y, Z2, color='blue')
    ax.scatter(0, 0, 0, 'z', 10, 'red')
    show()

draw_parallel_planes(0.6, 0.8, z_offset=2)
```

```python
def draw_fig():
    fig = figure()
    ax = Axes3D(fig)
    X = np.arange(-20, 20, 1)
    Y = np.arange(-10, 10, 1)
    X, Y = np.meshgrid(X, Y)
    Z = -X**2 - Y**2
    Z2 = np.cos(X/1.4)*10 + np.sin(Y/1.4)*10
    ax.plot_wireframe(X, Y, Z2, color='red')
    ax.plot_wireframe(X, Y, Z, color='green')
    ax.scatter(0, 0, 0, 'z', 50, 'red')
    show()
    
draw_fig()
```

```python
def polar_to_decart(r, angle):
    # angle in radians
    x = r * np.cos(angle)
    y = r * np.sin(angle)
    return(x, y)

polar_to_decart(1, np.pi)
```
